```python
import pandas as pd, numpy as np, requests
from bs4 import BeautifulSoup as BS
```

```python
url = 'https://dnd5e.wikidot.com/armor'
resp = requests.get(url)
```

```python
armor_tables = pd.read_html(resp.content)
armor_tables[0]['type'] = 'light'
armor_tables[1]['type'] = 'medium'
armor_tables[2]['type'] = 'heavy'
armor_tables[3]['type'] = 'shield'
```

```python
armor_tables[0]['Strength'] = 0
armor_tables[0].columns = ['Name','AC','Strength','Stealth','Weight','Cost','type']
armor_tables[0]['baseAC'] = armor_tables[0]['AC'].str.split('+',expand=True)[0]
armor_tables[0]['baseAC'] = armor_tables[0]['baseAC'].str.strip().astype(int)
armor_tables[0]['modAC'] = "DEX"
armor_tables[0]['Weight'] = armor_tables[0]['Weight'].str[:-4]
armor_tables[0]['Weight'] = armor_tables[0]['Weight'].astype(int)
armor_tables[0][['cost_amt','cost_unit']] = armor_tables[0]['Cost'].str.split(' ',expand=True)
armor_tables[0]['cost_amt'] = armor_tables[0]['cost_amt'].astype(int)
armor_tables[0]['Stealth'] = np.select(
    [armor_tables[0]['Stealth']=='Disadvantage',armor_tables[0]['Stealth']=='-'],
    [True,False]
)
armor_tables[0]['Stealth'] = armor_tables[0]['Stealth'].astype(bool)
armor_tables[0].drop(columns=['AC','Cost'], axis=1,inplace=True)
armor_tables[0].columns=['Name','minSTR','stealth_disadv','Weight','type','baseAC','modAC','cost_amt','cost_unit']
armor_tables[0]
```

```python
armor_tables[1]['Strength'] = 0
armor_tables[1]['Strength'] = armor_tables[1]['Strength'].astype(int)
armor_tables[1]['Stealth'] = np.select(
    [armor_tables[1]['Stealth']=='Disadvantage',armor_tables[1]['Stealth']=='-'],
    [True,False]
)
armor_tables[1]['Stealth'] = armor_tables[1]['Stealth'].astype(bool)
armor_tables[1].columns = ['Name','AC','minSTR','stealth_disadv','Weight','Cost','type']
armor_tables[1]['baseAC'] = armor_tables[1]['AC'].str.split('+',expand=True)[0]
armor_tables[1]['baseAC'] = armor_tables[1]['baseAC'].str.strip().astype(int)
armor_tables[1]['modAC'] = "DEX|2"
armor_tables[1]['Weight'] = armor_tables[1]['Weight'].str[:-4]
armor_tables[1]['Weight'] = armor_tables[1]['Weight'].astype(int)
armor_tables[1][['cost_amt','cost_unit']] = armor_tables[1]['Cost'].str.split(' ',expand=True)
armor_tables[1]['cost_amt'] = armor_tables[1]['cost_amt'].astype(int)
armor_tables[1].drop(columns=['AC','Cost'],axis=1,inplace=True)
armor_tables[1]
```

```python
# armor_tables[2].loc[armor_tables['Strength'].str.contains('-'),'Strength'] = 0
# armor_tables[2]['Stealth'] = True
armor_tables[2]
armor_tables[2].loc[armor_tables[2]['Strength'].str.contains('-'),'Strength'] = 'Str 0'
armor_tables[2]['Strength'] = armor_tables[2]['Strength'].str.split(' ',expand=True)[1].astype(int)
armor_tables[2]['Stealth'] = np.select(
    [armor_tables[2]['Stealth']=='Disadvantage',armor_tables[2]['Stealth']=='-'],
    [True,False]
)
armor_tables[2]['Stealth'] = armor_tables[2]['Stealth'].astype(bool)
armor_tables[2]['Weight'] = armor_tables[2]['Weight'].str[:-4]
armor_tables[2]['Weight'] = armor_tables[2]['Weight'].astype(int)
armor_tables[2][['cost_amt','cost_unit']] = armor_tables[2]['Cost'].str.split(' ',expand=True)
armor_tables[2]['cost_amt'] = armor_tables[2]['cost_amt'].str.replace(',','').astype(int)
# armor_tables[2]['cost_amt'] = armor_tables[2]['cost_amt'].astype(int)
armor_tables[2].columns = ['Name','baseAC','minSTR','stealth_disadv','Weight','Cost','type','cost_amt','cost_unit']
armor_tables[2].drop(columns=['Cost'],inplace=True)
armor_tables[2]['modAC'] = None
armor_tables[2]
```

```python
armor_tables[3].columns=['Name','baseAC','minSTR','stealth_disadv','Weight','Cost','type']
armor_tables[3].loc[armor_tables[3]['minSTR'].str.contains('-'),'minSTR'] = 0
armor_tables[3].loc[armor_tables[3]['stealth_disadv'].str.contains('-'),'stealth_disadv'] = False
armor_tables[3]['Weight'] = armor_tables[3]['Weight'].str[:-4].astype(int)
armor_tables[3][['cost_amt','cost_unit']] = armor_tables[3]['Cost'].str.split(' ',expand=True)
armor_tables[3]['cost_amt'] = armor_tables[3]['cost_amt'].astype(int)
armor_tables[3]['modAC'] = None
armor_tables[3].drop(columns=['Cost'],inplace=True)
armor_tables[3]
```

```python
armor_tables[4]['Category'] = np.select(
    [
        armor_tables[4]['Category'].str.contains('Light'),
        armor_tables[4]['Category'].str.contains('Medium'),
        armor_tables[4]['Category'].str.contains('Heavy'),
        armor_tables[4]['Category'].str.contains('Shield')
    ],[
        'light','medium','heavy','shield'
    ]
)
armor_tables[4]
```

```python
armor = pd.concat(armor_tables[0:4])
```

```python
armor
```

```python
armor = armor.merge(right=armor_tables[4],left_on='type',right_on=['Category'],how='left')
armor.drop(columns='Category',inplace=True,axis=1)
armor
```

```python
armor.to_csv('./data/armor.csv',index=False)
```
